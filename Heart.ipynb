```python
%pylab inline
import matplotlib.pyplot as plt
import math
import numpy as np
import warnings


def f(x):
    first = (math.cos(x) ** 0.5) * math.cos(200 * x) + (abs(x) ** 0.5) - 0.7
    second = pow(4 - x ** 2, 0.01)
    return first * second


warnings.filterwarnings('ignore')

xs = np.arange(-1.6, 1.6, 0.0001)
ys = [f(x) for x in xs]

plt.xlim([-3, 3])
plt.plot(xs, ys, color='red')
plt.show()
```
